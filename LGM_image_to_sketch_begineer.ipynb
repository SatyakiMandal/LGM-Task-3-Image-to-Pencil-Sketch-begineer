#**Image to Sketch**

**Import Libraries**

```python
import cv2
import matplotlib.pyplot as plt
```

**Read Photo**

```python
img=cv2.imread("pexels-hashtag-melvin-10556271.jpg")
```

**Display Using Matplotlib**

```python
plt.imshow(img)
plt.axis(False)
plt.show()
```

**Convert BGR to RGB**

```python
plt.imshow(img[:,:,::-1])
plt.axis(False)
plt.show()
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

**Convert Image to a Pencil Sketch**

*Step 1: Convert to Grey Image*

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

*Step 2: Invert Image*

```python
invert_img=cv2.bitwise_not(grey_img)
#invert_img=255-grey_img
```

*Step 3: Blur image*

```python
blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
```

*Step 4: Invert Blurred Image*

```python
invblur_img=cv2.bitwise_not(blur_img)
#invblur_img=255-blur_img
```

*Step 5: Sketch*

```python
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
```

*Step 6: Save Sketch*

```python
 cv2.imwrite('sketch.png', sketch_img)
```

*Step 7: Display sketch*

```python
img2 = cv2.imread("sketch.png")
plt.imshow(img2)
plt.axis(False)
plt.show()
```

**Original Image vs Sketch**

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
